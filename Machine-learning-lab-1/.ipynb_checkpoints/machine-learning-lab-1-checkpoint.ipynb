## Question 2
Create a load_boston_data(filepath: str) -> pd.DataFrame function.

This may require iterating through each line in the file until you read the data, at this point you will need to parse the data. Finally, return the data as a dataframe.

```python
def load_boston_data2(filepath:str) -> pd.DataFrame :
  with open(filepath, 'r') as csvfile:
      reader = csv.reader(csvfile, delimiter = ',')
      data = []
      for line in reader:
          data.append(line)

  df = pd.DataFrame(data[1:], columns=data[0])
  return df
filepath= '/home/joo/jnotebook/BostonHousing.csv'
df = load_boston_data(filepath)
print(df.head())
print(df.shape)
```

## Question 4
Create a function called lm, that takes an x, and y, and returns the random m and b variables in the linear equation:

```python
def lm(x,y) -> tuple[float, float]:

  m = np.random.rand()
  b = np.random.rand()

  return m,b
```

## Question 5
Using these m, b variables, create a housing price prediction for each row of data.

```python

def cal_y_pred(x,y):
  m, b = lm(x,y)
  y_pred = m * x + b
  return y_pred
x = df['rm']
y = df['medv']
y_pred = cal_y_pred(x,y)
print(y_pred[0:10])
```

## Question 6
Create a function mae that calculate the mean absolute error of the true house price value and the predicted value. What is the error?

```python
def mae(y,y_pred) -> float :
  return sum(abs(y - y_pred)) / len(y)

x = df['rm']
y = df['medv']
y_pred = cal_y_pred(x,y)
mae1 = mae(y,y_pred)
print(mae1)
```

## least-squares regression

```python
def lm_2(x,y) -> tuple[float, float]:

  m = sum((x - np.mean(x)) * (y - np.mean(y))) / sum((x - np.mean(x)) ** 2)
  b = np.mean(y) - m * np.mean(x)
  return m,b

def cal_y_pred_2(x,y):
  m, b = lm_2(x,y)
  y_pred = m * x + b
  return y_pred

x = df['rm']
y = df['medv']
y_pred = cal_y_pred_2(x,y)
print(y_pred[0:10])

plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.xlabel('rm')
plt.ylabel('medv')
plt.show()

mae2 = mae(y,y_pred)
print(f'mae2: {mae2}')
```

## gradient descent method 

```python
def lm_3(x,y,lr=0.01, epoch=1000 )-> tuple[float, float]:
  n=len(x)
  m = 0;
  b = 0;
  for _ in range(epoch):
    y_pred = m * x + b
    cost = (1/n) * sum((y - y_pred)**2)
    dm = (-2/n) * sum(x * (y - y_pred))
    db = (-2/n) * sum(y - y_pred)
    m = m - lr * dm
    b = b - lr * db
  return m,b

def cal_y_pred_3(x,y):
  m, b = lm_3(x,y)
  y_pred = m * x + b
  return y_pred

x = df['rm']
y = df['medv']
y_pred = cal_y_pred_3(x,y)
print(y_pred[0:10])

plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.xlabel('rm')
plt.ylabel('medv')
plt.show()


mae3 = mae(y,y_pred)
print(f'mae3: {mae3}')
```
